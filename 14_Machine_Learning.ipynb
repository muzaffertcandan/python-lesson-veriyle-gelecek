```python
# Standard Deviation
# Standart sapma, değerlerin ne kadar yayıldığını tanımlayan bir sayıdır.
# ortalama değerden ne kadar uzak olduğudur
speed = [86,87,88,86,87,85,86]

x = numpy.std(speed)
y= numpy.mean(speed)
print("x: ",x)
print("y: ",y)
```

```python
# example 2:
speed = [32,111,138,28,59,77,97]

x = numpy.std(speed)

print(x)
```

```python
# Variance
# varyans değerlerin ne kadar dağıldığını gösteren değerdir
# ve ortalama değerin her bir değerden çıkarılmasıyla bulunur
# varyansın karekökü standart sapmaya eşittir
# (32+111+138+28+59+77+97) / 7 = 77.4 ort değer
"""
 32 - 77.4 = -45.4
111 - 77.4 =  33.6
138 - 77.4 =  60.6
 28 - 77.4 = -49.4
 59 - 77.4 = -18.4
 77 - 77.4 = - 0.4
 97 - 77.4 =  19.6
 """
"""
(-45.4)2 = 2061.16
 (33.6)2 = 1128.96
 (60.6)2 = 3672.36
(-49.4)2 = 2440.36
(-18.4)2 =  338.56
(- 0.4)2 =    0.16
 (19.6)2 =  384.16
 """
# (2061.16+1128.96+3672.36+2440.36+338.56+0.16+384.16) / 7 = 1432.2

```

```python
#numpy ile varyans bulma:
speed = [32,111,138,28,59,77,97]

x = numpy.var(speed)

print(x)
```

```python
# bu değerlerin standart sapmasını bulmak için
# √1432.25 = 37.85
# standrat sapma σ ile gösterilir
# varyans σ nin karesidir
```

```python
# Percentile
# Percentile lar bir dizideki sayıların verdiğiniz yüzdeye göre daha küçük olduğunu gösterir
# örnek bir yaş dizisindeki %75 inin kaçtan küçük olduğunu percentile ile bulabiliriz
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = numpy.percentile(ages, 75)

print(x)
```

```python
# insanların %90 ının daha küçük olduğu yaş:
```

```python
x = numpy.percentile(ages, 90)

print(x)
```

```python
# Data Distribution
# Gerçek hayatta veri toplamak zor olabilir bunun için
# Random veriler üretiriz
import numpy
# 0.0 dan 5.0 a kadar 250 adet random float veriler ürettik
x = numpy.random.uniform(0.0, 5.0, 250)

print(x)
```

```python
# histogram verilerimizi görselleştirmek için gerekir
import matplotlib.pyplot as plt
# ilk sütun 0 ile 1 arasındaki sayıları temsil ediyor
x = numpy.random.uniform(0.0, 5.0, 250)

plt.hist(x, 15)
plt.show()
```

```python
#veriler istenildiği kadar büyük olabilir
x = numpy.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 100)
plt.show()
```

```python
# Veriler tek bir noktada yoğunluklu da olabilir ve bu tarz veri dağılımna
# Normal Data Distribution veya Gaussian data distribution denir
# ortalama değer 5 ve standart sapma 1
x = numpy.random.normal(5.0, 1.0, 100000)
# 100 satırlı bir histogram çizmek için:
plt.hist(x, 8 )
plt.show()
```

```python
# Regresyon 2 değişken arasındaki ilişkiyi bulmak için kullanılır
# makine öğrenmesi ve istatistiksel modellemeler gelecekte bize yardımcı olması için tahminlerde bulunurlar.
# Linear Regression veri noktaları arasındaki ilişkileri kullanarak düz bir çizgi çizer
# bu çizgi gelecek tahminleri yapmak için kullanılır
```

```python

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
